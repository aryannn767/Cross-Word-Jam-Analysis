```python
import pandas as pd
from google_play_scraper import reviews
```

### Fetching App Review Data From Play Store

```python
result, _ = reviews('in.playsimple.tripcross', lang='en', country='in', count=2000)

df = pd.DataFrame(result)
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
streak_keywords = [
    'streak', 'daily', 'every day', 'habit',
    'login', 'day', 'reward', 'bonus', 'keep playing'
]

df['mentions_streak'] = df['content'].str.contains('|'.join(streak_keywords))
```

```python
df.info()
```

```python
df['mentions_streak'].value_counts()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

df['sentiment_score'] = df['content'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
```

```python
df.info()
```

```python
## KPI Calculation

## Overall Sentiment
df['sentiment_score'].mean() ## 0.0341

## Overall user sentiment is barely positive, indicating that the user experience is average and not strongly delightful
```

```python
streak_avg = df[df['mentions_streak']]['sentiment_score'].mean()
print(streak_avg)
```

```python
non_streak_avg = df[~df['mentions_streak']]['sentiment_score'].mean()
print(non_streak_avg)
```

```python
delta = streak_avg - non_streak_avg
print(delta)
```

```python
df['mentions_streak'].mean()
```

```python
df.groupby('mentions_streak')['score'].mean()
```

```python
## Visualization


sentiment_comparison = df.groupby('mentions_streak')['sentiment_score'].mean()
print(sentiment_comparison)
```

```python
## Sentiment Comparison

import matplotlib.pyplot as plt

sentiment_comparison.plot(kind='bar')

plt.title("Sentiment: Streak vs Non-Streak Users")
plt.xlabel("Mentions Streak")
plt.ylabel("Average Sentiment")
plt.xticks([0,1], ['Non-Streak', 'Streak'], rotation=0)

plt.show()
```

```python
## Rating Distribution

df['score'].value_counts().sort_index().plot(kind='bar')

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['at'])
df['month'] = df['at'].dt.to_period('M')
```

```python
## Sentiment Trend Over Time

monthly_sentiment = df.groupby('month')['sentiment_score'].mean()
print(monthly_sentiment)
```

```python
monthly_sentiment.plot()

plt.title("Sentiment Over Time")
plt.xlabel("Month")
plt.ylabel("Average Sentiment")

plt.show()
```

```python
## Compare Trend for Streak vs Non-Streak

trend = df.groupby(['month', 'mentions_streak'])['sentiment_score'].mean().unstack()

trend.plot()

plt.title("Sentiment Trend: Streak vs Non-Streak")
plt.xlabel("Month")
plt.ylabel("Sentiment")

plt.show()
```
